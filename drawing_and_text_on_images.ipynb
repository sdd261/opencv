```python
import cv2
import numpy as np
```

```python
img = np.zeros((512,512,3))
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
cv2.rectangle(img,pt1=(100,100),pt2=(300,300),color=(255,0,0),thickness=2) #thickness (-1) means fill in the place
cv2_imshow(img)
```

```python
cv2.circle(img,center=(256,256),radius=100,color=(0,255,0),thickness=2)
cv2_imshow(img)
```

```python
cv2.line(img,pt1=(0,0),pt2=(512,512),color=(0,0,255),thickness=2)
cv2_imshow(img)
```

```python
cv2.putText(img,org=(400,400),fontScale=4,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,fontFace=cv2.FONT_HERSHEY_COMPLEX,text="Hello")
cv2_imshow(img)
```
